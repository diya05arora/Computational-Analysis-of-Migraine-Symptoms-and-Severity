Univariate Analysis

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Duration'], kde=True, bins=30)
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
# only 1, 2 and 3
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Frequency'], kde=True, bins=30)
plt.title('Distribution of Frequency')
plt.xlabel('Frequency')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Frequency'], kde=True, bins=30)
plt.title('Count of Different Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

Bivariate Analysis

```python
# Duration vs Age

plt.figure(figsize=(12, 6))
sns.histplot(x=df['Duration'], y=df['Age'], cbar="winter")
plt.title('Duration vs Age')
plt.xlabel('Duration')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(x=df['Frequency'], y=df['Intensity'], cbar="winter")
plt.title('Intensity vs Frequency')
plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(df['Character'])
plt.title('Count of Different Characters')
plt.xlabel('Character')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
